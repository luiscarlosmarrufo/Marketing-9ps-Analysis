```python
import pandas as pd
import json 
import seaborn as sns
import matplotlib.pyplot as plt
import plotly as py
```

```python
x_file = "Data_X.ndjson"

x_data = []
with open(x_file, 'r', encoding='utf-8') as file:
    for line in file:
        x_data.append(json.loads(line))

df_x = pd.DataFrame(x_data)
```

```python
# Explorar la estructura de un registro en la columna 'data'
sample_data = df_x['data'].iloc[0]
print("Estructura de 'data':")
for key, value in sample_data.items():
    print(f"{key}: {type(value)}")
```

```python
# Expandir la columna 'data' en un DataFrame
data_expanded = pd.json_normalize(df_x['data'])

# Unir el DataFrame expandido con el original
df_x_expanded = pd.concat([df_x.drop(columns=['data']), data_expanded], axis=1)

# Mostrar las primeras filas del DataFrame expandido
df_x_expanded.head()
```

```python
df_x_expanded = df_x_expanded.dropna(axis=1, how='any')
```

```python
# Seleccionar columnas relevantes
columns_to_keep = [
    'item_id',        # ID del video
    'legacy.full_text',           # Texto del comentario
    'legacy.created_at',    # Fecha de creacin del comentario
    'legacy.favorite_count'      # Likes del comentario
]
```

```python
# Crear un nuevo DataFrame con las columnas seleccionadas
df_x_selected = df_x_expanded[columns_to_keep].copy()
```

```python
# Renombrar columnas para mayor claridad
df_x_selected.rename(columns={
    'item_id': 'id',
    'legacy.full_text': 'text',
    'legacy.created_at': 'date',
    'legacy.favorite_count': 'likes'
}, inplace=True)
```

```python
# Mostrar las primeras filas para verificar
print("Datos seleccionados de TikTok:")
df_x_selected.head(10)
```

```python
# Filter rows to keep only those containing "starbucks" in the "text" column
df_x_selected = df_x_selected[df_x_selected['text'].str.contains('starbucks', case=False, na=False)]

# Mostrar las primeras filas para verificar
print("Datos seleccionados de TikTok (solo con 'starbucks'):")
df_x_selected.head(50)
```

```python
from langdetect import detect
from deep_translator import GoogleTranslator
import pandas as pd

# Function to detect language
def detect_language(text):
    try:
        return detect(str(text))
    except:
        return 'unknown'

# Function to translate text to English
def translate_to_english(text, source_lang):
    if pd.isna(text) or source_lang == 'en' or source_lang == 'unknown':
        return text
    try:
        # If source language is known, use it for better translation
        translator = GoogleTranslator(source=source_lang if source_lang != 'unknown' else 'auto', target='en')
        return translator.translate(str(text))
    except:
        return text
```

```python
# Add language detection column
df_x_selected['detected_language'] = df_x_selected['text'].apply(detect_language)

# Translate the text column to English
df_x_selected['text_english'] = df_x_selected.apply(
    lambda row: translate_to_english(row['text'], row['detected_language']), axis=1
)

# Display the results
print("Data with language detection and translation:")
df_x_selected[['text', 'detected_language', 'text_english']].head(10)
```
